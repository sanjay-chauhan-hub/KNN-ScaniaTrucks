## Create Pandas datasets from raw Scania Truck Dataset

**Get the data from https://archive.ics.uci.edu/ml/machine-learning-databases/00421/.**

**Convert datasets to Pandas dataframe and create new CSV files**

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("aps_failure_training_set.csv", sep = "delimiter", header = None)
```

```python
data
```

```python
for i in range(15):
    data.drop(i, inplace = True)
```

```python
data
```

```python
columnNames = data.loc[14]
```

```python
print(columnNames)
print(type(columnNames))
```

```python
columnNamesSplit = columnNames.str.split(",")
```

```python
print(type(columnNamesSplit))
print(columnNamesSplit)
```

```python
columnNames = []
for j in columnNamesSplit:
    for l in j:
        columnNames.append(l)
```

```python
rows = []
for i in range(60000):
    subRows = []
    row_i = data.loc[i + 15]
    row_i_Split = row_i.str.split(",")
    for j in row_i_Split:
        for l in j:
            subRows.append(l)
    rows.append(subRows)
```

```python
df = pd.DataFrame(rows, columns=columnNames)
```

```python
df
```

```python
df.to_csv("ScaniaTrucksTrainingSet.csv")
```
